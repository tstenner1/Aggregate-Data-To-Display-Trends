```python
# Dependencies and Setup
import pandas as pd
import numpy as np

schools_complete = "Resources/schools_complete.csv"
students_complete = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
schools_df = pd.read_csv(schools_complete, encoding = "utf-8")
schools_df.head()

students_df = pd.read_csv(students_complete, encoding = "utf-8")
students_df.head()

# Combine the data into a single dataset  
merge_table = pd.merge(students_df, schools_df, how="outer", on=["school_name", "school_name"])
merge_table.head()
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
#Calculate the total number of schools 
schools_total = schools_df["school_name"].count()


#Calculate the total number of students 
students_total = students_df["Student ID"].count()


#Calculate the total budget 
total_budget = schools_df["budget"].sum()


#Calculate the average math score 
math_average = students_df["math_score"].mean()
math_average_rounded = round(math_average, 2)

#Calculate the average reading score 
reading_average = students_df ["reading_score"].mean()
reading_average_rounded = round(reading_average, 2)

#Calculate the percentage of students with a passing math score (70 or greater)
passing_math_df = students_df.loc[students_df["math_score"]>= 70, :]
passing_math_df.head()

passing_math_total = (passing_math_df["Student ID"].count()/ students_total) * 100
passing_math_rounded = round(passing_math_total, 2)



#Calculate the percentage of students with a passing reading score (70 or greater)
passing_reading_df = students_df.loc[students_df["reading_score"]>= 70, :]
passing_reading_df.head()

passing_reading = (passing_reading_df["Student ID"].count()/students_total)*100
passing_reading_rounded = round(passing_reading,2)

#Calculate the percentage of students who passed math and reading (% Overall Passing)
overall_pass = (passing_math_rounded + passing_reading_rounded)/2

#Create a dataframe to hold the above results
results_df = pd.DataFrame({"Total Schools": [schools_total],
                           "Total Students": [students_total],
                           "Total Budget": [total_budget],
                           "Average Math Score": [math_average_rounded],
                           "Average Reading Score": [reading_average_rounded],
                           "% Passing Math": [passing_math_rounded],
                           "% Passing Reading": [passing_reading_rounded],
                           "Overall Passing Rate": [overall_pass]})
results_df
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Create a dataframe to hold the above results

```python
#Individual schools average reading scores 
average_reading_score = students_df.groupby('school_name')['reading_score'].mean()

average_reading_score_df = pd.DataFrame({'school_name':average_reading_score.index, 'average_reading_score':average_reading_score.values})

#Individual schools average math scores 

average_math_score = students_df.groupby('school_name')['math_score'].mean()

average_math_score_df = pd.DataFrame({'school_name':average_math_score.index, 'average_math_score':average_math_score.values})


#Students that are passing math class, by school 
count_passing_math = passing_math_df.groupby('school_name')['math_score'].count()


count_passing_math_df = pd.DataFrame({'school_name':count_passing_math.index, 'count_passing_math':count_passing_math.values})


#Students passing reading class, by school 
count_passing_reading = passing_reading_df.groupby('school_name')['reading_score'].count()


count_passing_reading_df = pd.DataFrame({'school_name':count_passing_reading.index, 'count_passing_reading':count_passing_reading.values})


#Outer join schools with the average reading scores  
schools_join_averageread_df = pd.merge(schools_df, average_reading_score_df, on="school_name", how="outer")
schools_join_averageread_df.head()

#Outer join schools with the average math scores  
schools_join_averagemath_df = pd.merge(schools_join_averageread_df, average_math_score_df, on="school_name", how="outer")
schools_join_averagemath_df.head()

#Join schools count of students passing reading by each individual school 
schools_join_countread_df = pd.merge(schools_join_averagemath_df, count_passing_reading_df, on="school_name", how="outer")
schools_join_countread_df.head()

#Join Schools with Count of Students Passing Math by School
schools_join_countmath_df = pd.merge(schools_join_countread_df, count_passing_math_df, on="school_name", how="outer")
schools_join_countmath_df.head()

#Percent of Students Passing Math by School
schools_join_countmath_df['percent_passing_math'] = (schools_join_countmath_df["count_passing_math"]/ schools_join_countmath_df["size"]) * 100


#Percent of Student Passing Reading by School
schools_join_countmath_df['percent_passing_reading'] = (schools_join_countmath_df["count_passing_reading"]/ schools_join_countmath_df["size"]) * 100


#Overall Passing Rate by School
schools_join_countmath_df['Overall_passing_rate'] = (schools_join_countmath_df["percent_passing_reading"] + schools_join_countmath_df["percent_passing_math"])/ 2
schools_join_countmath_df.head()

#Budget per Student by School
schools_join_countmath_df['Per_student_budget'] = (schools_join_countmath_df["budget"] / schools_join_countmath_df["size"])
schools_join_countmath_df.head()

#Rename so it is less to type later in the code 
schools_summary_noformat_fortype_df = schools_join_countmath_df
schools_summary_noformat_fortype_df.head()

#Renaming columns 
school_rename_df = schools_summary_noformat_df.rename(columns={"school_name":"School Name", 
                                                             "type":"School Type",
                                                             "size": "Total Students", 
                                                             "budget": "Total School Budget", 
                                                             "Per_student_budget": "Per Student Budget", 
                                                             "avg_math_score": "Average Math Score",
                                                             "avg_reading_score": "Average Reading Score",
                                                             "percent_passing_math": "% Passing Math",
                                                             "percent_passing_reading": "% Passing Reading",
                                                             "Overall_passing_rate": "% Overall Passing Rate"})

school_rename_df.head()

#Reformating columns
school_rename_df["Total School Budget"] = school_rename_df["Total School Budget"].map("${:,.2f}".format)
school_rename_df["Per Student Budget"] = school_rename_df["Per Student Budget"].map("${:,.2f}".format)
school_rename_df.head()

#final dataframe
school_summary_df = school_rename_df.loc[:, ['School Name', 'School Type', 'Total Students',
       'Total School Budget', 'Per Student Budget','Average Reading Score', 'Average Math Score',
       '% Passing Math','% Passing Reading', '% Overall Passing Rate']]

school_summary_df
```
